```python
def rnd(N=100000):
    return random.random(N)
```

```python
def generate_s_xyz():
    phi = 2 * np.pi * rnd()
    r = np.cbrt(rnd())
    
    cos_t = 2 * rnd() - 1    
    sin_t = np.sin(np.arccos(cos_t))
    cos_p = np.cos(phi)
    sin_p = np.sin(phi)
    
    sx = r * sin_t * cos_p 
    sy = r * sin_t * sin_p
    sz = r * cos_t
    
    return sx, sy, sz


def calculate_ab(R, M, m):
    return np.sqrt(1 + np.square(R / M)),R / np.sqrt(M * M - 4 * m * m)


def calculate_p_xyz(s, a, b):
    sx, sy, sz = s
    p1x = a * sx + b
    p2x = a * sx - b
    return (p1x, sy, sz), (p2x, -sy, -sz)


def dphi_gen(a, b):
    sx = (2 * np.random.random(N) - 1)
    sy = np.sqrt(1 - sx * sx)
    sz = 0
    
    p1x = a * sx + b
    p2x = a * sx - b
    
    tgf1 = sy / p1x
    tgf2 = - sy / p2x
    
    tgdf = (tgf1 - tgf2) / (1 + tgf1 * tgf2)
    return np.arctan(tgdf)

def dtheta_gen(a, b):
    sx = (2 * np.random.random(N) - 1)
    sy = 0
    sz = np.sqrt(1 - sx * sx) * np.random.choice([-1,1], N)
    
    p1x = a * sx + b
    p2x = a * sx - b
    
    theta1 = np.arctan(sz / (sy * sy + p1x * p1x))
    theta2 = np.arctan(-sz / (sy * sy + p2x * p2x))
    
    dn = 0.5 * np.log(np.tan((np.pi / 2 - theta1) / 2)/ (np.tan((np.pi / 2 - theta2) / 2)))
    return dn
```

```python
p = figure(tools="crosshair,pan,zoom_in,zoom_out")

def plot_s_xy():
    sx, sy, _ = generate_s_xyz()
    p.scatter(sx, sy, size=((-2,2),(-2,2)),radius=.01,fill_alpha=0.6, line_color=None)
    p.scatter(0, 0,color="#ff0000", radius=.03,fill_alpha=0.6, line_color=None)
    show(p)
    
def plot_p_yx(a, b):
    p1, _ = calculate_p_xyz(generate_s_xyz(), a, b)
    px, py, _ = p1
    p.scatter(py, px, radius=.01,fill_alpha=0.3, line_color=None)
    p.scatter(0, 0,color="#ff0000", radius=.03,fill_alpha=0.6, line_color=None)
    show(p)

def plot_p_zx(a, b):
    p1, _ = calculate_p_xyz(generate_s_xyz(), a, b)
    px, _, pz = p1
    p.scatter(pz, px, radius=.01,fill_alpha=0.3, line_color=None)
    p.scatter(0, 0,color="#ff0000", radius=.03,fill_alpha=0.6, line_color=None)
    show(p)
```
